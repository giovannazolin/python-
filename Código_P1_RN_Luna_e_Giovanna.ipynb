```python
#Função da MLP
class MLP(object):
    # Dicionário do python para armazenar o modelo da MLP
	mod = {}
	d_h_b = 0
	d_o_b = 0
    # Construtor da classe
	def __init__(self):
		# Iniciando MLP
		return
    # Função de Ativação dos neuronios da MLP
	def fneu(net):
		return (1 / (1 + np.exp(-net)))

    # Função para calcular a derivada
	def cal_d(f_net):
		return (f_net * (1 - f_net))

    # Função que inicializa a arquitetura da MLP baseado no problema especifico
	def arch(self,input_lenght=256,hidden_lenght=20,output_lenght=10,fneu=fneu,cal_d=cal_d):
		self.mod['input_lenght'] = input_lenght
		self.mod['hidden_lenght'] = hidden_lenght
		self.mod['output_lenght'] = output_lenght
		self.mod['hidden_layer'] = np.random.uniform(-0.5,+0.5,(hidden_lenght,input_lenght+1))
		self.mod['output_layer'] = np.random.uniform(-0.5,+0.5,(output_lenght,hidden_lenght+1))
		self.mod['fneu'] = fneu
		self.mod['cal_d'] = cal_d
        # Realiza o forward do algoritmo, onde calcula o output final a partir dos pesos atuais
	def forward(self, X):
		#Retirando os valores do modelo
		hidden = self.mod['hidden_layer'] 
		output = self.mod['output_layer']
		#Adicionando o 1 para a multiplicação
		X = np.concatenate((X,np.array([1])))


		
		#CAMADA ESCONDIDA
		neu_h = np.matmul(hidden,X)
		fneu_h = self.mod['fneu'](neu_h)
		cal_d_h = self.mod['cal_d'](fneu_h)
		
		#CAMADA SAÍDA
		fneu_h_c = np.concatenate((fneu_h,np.array([1])))
		net_o = np.matmul(output,fneu_h_c)
		fneu_o = self.mod['fneu'](net_o)
		cal_d_o = self.mod['cal_d'](fneu_o)
        
		return{
			"fneu_h": fneu_h,
			"cal_d_h":cal_d_h,
			"fneu_o": fneu_o,
			"cal_d_o":cal_d_o
		}

    # Realiza o treinamento da rede utilizando backpropagation com regra delta
	def backpropagation(self,X,Y,eta=0.5,alpha=0.5,max_error=0.000001,max_iter=500):
		counter = 0
		t_e = 2*max_error

		# Treinamento ocorre enquanto o erro for maior que o aceitável ou o número maximo de iterações não tiver sido atingido
		while t_e > max_error and counter < max_iter:
			t_e = 0

			for i in range(0,X.shape[0]):
				x_i = X[i,:]
				y_i = Y[i]

				#forward
				fw = self.forward(x_i)

				#erro
				error_o_k = (y_i-fw['fneu_o'])
				
				t_e = t_e + np.sum(error_o_k*error_o_k)

				#backpropagation / calculo das derivadas
				delta_out = error_o_k*fw['cal_d_o']
				d_o = np.multiply(np.array([-2*delta_out]).T,np.concatenate((fw['fneu_h'],np.array([1]))))

				delta_h = np.matmul(np.array([delta_out]),self.mod['output_layer'][:,0:self.mod['hidden_lenght']])
				d_h = delta_h * (np.multiply(-2*fw['cal_d_h'],np.array([np.concatenate((x_i,np.array([1])))]).T))


				# Atualização dos pesos
				self.mod['output_layer'] = self.mod['output_layer'] - eta*d_o - alpha*self.d_o_b
				if counter == 0:
					self.mod['hidden_layer'] = self.mod['hidden_layer'] - np.reshape(np.array([eta*d_h]).T,(self.mod['hidden_lenght'],x_i.size+1))
				else:
					self.mod['hidden_layer'] = self.mod['hidden_layer'] - np.reshape(np.array([eta*d_h]).T,(self.mod['hidden_lenght'],x_i.size+1)) - np.reshape(np.array([alpha*self.d_h_b]).T,(self.mod['hidden_lenght'],x_i.size+1))
				

				self.d_o_b = d_o
				self.d_h_b = d_h

			# Término da iteração do treinamento
			t_e = t_e/X.shape[0]
			counter = counter+1
			if (counter % 100) == 0:
				print("Iter:",counter," Error:",t_e)

		return

	def run(self,X,Y,task,size=20,eta=0.1,alpha=0.5,max_iter=500,train_size=0.7,threshold=0.000001):
		ale = random.sample(range(0,X.shape[0]),np.floor(train_size*X.shape[0]).astype(np.int))
		ale_esq = diff(range(0,X.shape[0]),ale)
		# Base de Treino
		train_set = X[ale,:]
		Y=np.array(Y)
		train_classes = Y[ale,:]

		# Base de Teste
		test_set = X[ale_esq,:]
		test_classes = Y[ale_esq,:]

		self.arch(input_lenght=X.shape[1],hidden_lenght=size,output_lenght=Y.shape[1])
		print('MLP criada\nIniciando Treinamento')
		self.backpropagation(train_set,train_classes,eta=eta,alpha=alpha,max_error=threshold,max_iter=max_iter)
		print('Rede Neural Treinada\nIniciando Teste')

		correct = 0
		sqerror = 0
		for i in range(0,test_set.shape[0]):
			x_i = test_set[i]
			y_i = test_classes[i]

			if task == 'C':
				y_hat_i = np.round(self.forward(x_i)['fneu_o'])
			if task == 'R':
				y_hat_i = self.forward(x_i)['fneu_o']
			
			error = y_i - y_hat_i
			if (np.sum((error)**2) == 0):
				correct = correct + 1
			sqerror = sqerror + np.sum(error*error)

			pass

		print('Rede Neural Testada')
		accuracy = correct/test_set.shape[0]
		sqerror = sqerror/test_set.shape[0]
		
		return {
			"accuracy": accuracy,
			"error": sqerror    
		}
	#FIM DA MLP    
```

```python
def diff(first, second):
        second = set(second)
        return [item for item in first if item not in second]
```

```python
#Função para aplicação nos dados
def digitos(X,eta=0.1,alpha=0,max_iter=500,size=20,train_size=0.7):
    X = np.array(X)
    X = X.astype(np.float)
    # Normalizando X
    for i in range(X.shape[1]):
        X[:,i] = (X[:,i] - np.amin(X[:,i])) / (np.amax(X[:,i]) - np.amin(X[:,i]))
    print('\nProcessando Base de Dígitos')
    mlp = MLP()
    return mlp.run(X,Y,'C',eta=eta,alpha=alpha,size = size,max_iter=max_iter,train_size=train_size)
```
